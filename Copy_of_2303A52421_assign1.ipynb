Perform Mathematical Operations on these datasets - Multiplication, Division, Poweroff

```python
import numpy as np

lst1 = [1, 2, 3]
array1 = np.array(lst1)

lst2 = [4, 5, 6]
array2 = np.array(lst2)

print("array2 multiplied by array1: ", array2 * array1)
print("array2 divided by array1: ", array2 / array1)
print("array2 raised to the power of array1: ", array2 ** array1)
```

```python
import numpy as np

lst1 = [1, 2, 3]
array1 = np.array(lst1)
print("Sine: ", np.sin(array1))
print("Natural logarithm: ", np.log(array1))
print("Base-10 logarithm: ", np.log10(array1))
print("Base-2 logarithm: ", np.log2(array1))
print("Exponential: ", np.exp(array1))
```

Part-02: Lab Task is to visualize the datasets or data using Matplotlib builtin function Graphs
Generate Scatterplot
Let's suppose we collected age (in years), height (in cm), and weight (in kilograms) information from few patients visitng a clinic.

```python
import pandas as pd

people = ['Ann','Brandon','Chen','David','Emily','Farook',
          'Gagan','Hamish','Imran','Julio','Katherine','Lily']
age = [22,14,35,46,39,19,27,54,7,42,49,17]
weight = [56,34,78,69,72,63,74,67,19,66,84,49]
height = [163,136,171,166,175,169,174,158,106,172,157,159]

# Create a pandas DataFrame
data = {'Name': people, 'Age': age, 'Weight': weight, 'Height': height}
df = pd.DataFrame(data)

print(df)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight, height)
plt.title("Relationship between Weight and Height of Patients")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of Patients in Kgs")
plt.show()
```

Generate Bar Plot

```python
import matplotlib.pyplot as plt

people = ['Ann', 'Brandon', 'Chen', 'David', 'Emily', 'Farook',
          'Gagan', 'Hamish', 'Imran', 'Julio', 'Katherine', 'Lily']
plt.figure(figsize=(12, 4))
plt.title("People's weight in kgs", fontsize=16, fontstyle='italic')
plt.bar(x=people, height=weight, width=0.6, color='green', edgecolor='k', alpha=0.6)
plt.xlabel("People", fontsize=15)
plt.xticks(fontsize=14, rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("Weight (in kgs)", fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
weight = [56,34,78,69,72,63,74,67,19,66,84,49]
plt.figure(figsize=(7, 5))
plt.hist(weight, color='red', edgecolor='k', alpha=0.75, bins=5)
plt.title("Histogram of patient weight", fontsize=18)
plt.xlabel("Weight in kgs", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
